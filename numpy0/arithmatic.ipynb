**simple arithmatic operations**

```python
# import numpy as np
import numpy as np
```

```python
arr = np.array([1,2,3])
print(arr + 1)
print(arr * 2)
print(arr - 3)
print(arr / 4)
print(arr ** 2)
print(arr % 2)
```

**Vectorized Operation**

```python
arr = np.array ([1.2, 2.3, 3.1, 4.2])
print(np.sqrt(arr))

print(np.round(arr))
```

**Elementwise Operation**

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
print(arr1 // arr2) #Floor Division
print(arr1 % arr2) 
print(arr1 ** arr2) 
print()
print(arr1 @ arr2) #Dot Product
print(arr1.dot(arr2)) #Dot Product
```

**Comparison Operators**

```python
marks = np.array([70,55,60, 90, 34, 80])
print(marks >= 80)
print(marks == 90)
print(marks != 80)
```

```python
# scalar + array
a = np.array([1,2,3])
b = 5

print(a+b)
```

```python
# 1D + 2D

a = np.array([
    [1,2,3],
    [4,5,6]
])
b = np.array([10, 20 ,30])

print(a.shape)
print(b.shape) # (3,) is considered as (1,3)

print()
print(a+b)        # array b considered as : b =( [ 10, 20 , 30 ]
                #                              [ 10, 20 , 30]  )
```

```python
#column vector + row vector
x = np.array([
    [1],
    [2],            # shape (3,1)
    [3]
])

y = np.array([10, 20, 30])  # shape (3,) considered as (1,3)

"""
x considered as : 
x = [
    [1, 1, 1],
    [2, 2, 2],          
    [3, 3, 3]
]

y considered  as : 
y =[
    [10, 20 ,30],
    [10, 20, 30],
    [10, 20, 30]
]
"""

print(x+y)
```

```python
# different dimention
a = np.array([1, 2, 3]) # a is : (3,) considered as (1,3)

b = np.array([      # b is (2,1)
    [10],
    [20]
])
"""
[1,2,3]
[1,2,3]
[1,2,3]

    +
[10, 10, 10]
[20, 20, 20]

"""
print(a+b)
```
