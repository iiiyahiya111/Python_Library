**simple arithmatic operations**

```python
# import numpy as np
import numpy as np
```

```python
arr = np.array([1,2,3])
print(arr + 1)
print(arr * 2)
print(arr - 3)
print(arr / 4)
print(arr ** 2)
print(arr % 2)
```

**Vectorized Operation**

```python
arr = np.array ([1.2, 2.3, 3.1, 4.2])
print(np.sqrt(arr))

print(np.round(arr))
```

**Elementwise Operation**

```python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
print(arr1 // arr2) #Floor Division
print(arr1 % arr2) 
print(arr1 ** arr2) 
print()
print(arr1 @ arr2) #Dot Product
print(arr1.dot(arr2)) #Dot Product
```

**Comparison Operators**

```python
marks = np.array([70,55,60, 90, 34, 80])
print(marks >= 80)
print(marks == 90)
print(marks != 80)
```
