**Random Number**

```python
import numpy as np

rn = np.random.default_rng()
print(rn.integers(1,10,3)) # integer function also can be written as integer(low=1, high=10, size=3)
print()
print(rn.integers(1,10,(2,2))) # random 2D array
```

```python
# using uniform method

import numpy as np;
print(np.random.uniform()) # random between 0 and 1
print(np.random.uniform(1,10)) # random between 1 and 10
print(np.random.uniform(1,10,3))
print()
print(np.random.uniform(1,10,(2,2)))
```

```python
print(np.random.uniform(-1,1))
```

```python
np.random.seed(1)
print(np.random.uniform(1,10)) # generate same number for specific seed
```

**shuffle** , **choice**

```python
import numpy as np
rng = np.random.default_rng()
arr = np.array([1,2,3,4,5])
rng.shuffle(arr)
print(arr)
```

```python
rng = np.random.default_rng()

arr = np.array(['apple', 'banana', 'orange'])

arr1 = rng.choice(arr)
print(arr1)

arr2 = rng.choice(arr, size = 2)
print(arr2)
```

```python
arr3 = rng.choice(arr, size = (2,2))
print(arr3)
```
