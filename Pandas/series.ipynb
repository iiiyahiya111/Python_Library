# Series

**using list**

```python
import pandas as pd

data = [100, 200, 300, 400] # this is a list

series = pd.Series(data)
print(series)
```

```python
data = [100, 200, 300, 400] # this is a list

series = pd.Series(data, index=['room1', 'room2', 'room3', 'room4'])
print(series)
```

**Access specific element using labels**

```python
data = [100, 200, 300, 400] # this is a list

series = pd.Series(data, index=['room1', 'room2', 'room3', 'room4'])
print(series.loc['room2']) # Access specific element using labels
print(series.iloc[2]) # Access specific element using integer location
print()
print(series[series > 200]) # Access elements based on condition#
```

**using dictionary**

```python
student = {"std1": 20, "std2": 21, "std3": 22}

series = pd.Series(student)
print(series)
```

```python
# update values 
series.loc['std1'] += 5
print(series)
```

```python
print(series[series > 21])
```
