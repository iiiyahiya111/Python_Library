# DataFrame

```python
import pandas as pd
```

```python
student = {
    "name" : ["yahiya", "abid" , "ali"],
    "roll" : [21, 22, 23],
    "dept" : ["cse", "eee", "ice"]
}

print(student)
print()

df = pd.DataFrame(student)
print(df)
```

```python
# custom index
df = pd.DataFrame(student)
print(df)
```

```python
print(df.loc[1])
```

**Add a new column to the DataFrame**

```python
df["age"] = [23, 23, 25]
print(df)
```

**Add a new column to the DataFrame**

```python
"""
in pandas , a plain dics is treated as columns, now rows. 
to create a single row dataframe we need to wrape the value in list or need tp use a list of dics
"""
```

```python
# using list of dics
new_dics = [{
    "name" : 'aisf',
    "roll" : 24,
    "dept" : 'eee',
    "age" : 22,
}]
new_row = pd.DataFrame(new_dics)
print(new_row)
```

```python
# using wraped valued dictionary
new_dics = {
    "name" : ['aisf'],
    "roll" : [24],
    "dept" : ['eee'],
    "age" : [22],
}
new_row = pd.DataFrame(new_dics)
print(new_row)
```

```python
df = pd.concat([df, new_row])
print(df)
```

**Add more than one row in data frame**

```python
new_rows = [
    {"name" : 'ajax', "roll" : 11, "dept" : "cse", "age" : 20},
    {"name" : 'amin', "roll" : 12, "dept" : "cse", "age" : 21},
]
new_rows = pd.DataFrame(new_rows)
df= pd.concat([df, new_rows], ignore_index=True)
print(df)
```
