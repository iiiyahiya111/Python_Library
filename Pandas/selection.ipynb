```python
import pandas as pd
```

**Selection by Column**

```python
df = pd.read_csv("students_large_dataset.csv")
print(df["name"])

#print(df["name"].to_string())  # for printing all the data 

print(df["department"])
```

```python
# multiple columns
print(df[["name", "department"]])
```

**selection by rows**

```python
print(df.loc[0])
```

```python
# set a column as index
df = pd.read_csv("students_large_dataset.csv", index_col="student_id")
print(df)
```

```python
# now access by new index (row wise)
print(df.loc[1180])
```

**Row and Column**

```python
print(df.loc[1177 : 1180, ["department", "cgpa"]])
```

```python
# integer indexing
print(df.iloc[0:10 , 0:2])
```

```python
print(df.iloc[::-1]) # reverse the data
```

```python
print(df.iloc[[3,2,1], [0,3]])  
```
